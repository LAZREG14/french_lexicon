```python
import json
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

def ecrire_json(chemin, contenu):
    w    = open(chemin, "w", encoding="utf-8")
    w.write(json.dumps(contenu, indent=2, ensure_ascii=False))
    w.close()
```

```python
# Creating a Service object
chrome_service = webdriver.chrome.service.Service(ChromeDriverManager().install())

# Configuring the path of the ChromeDriver executable
webdriver.chrome.service.executable_path = chrome_service

# Creating an instance of the Chrome browser
driver = webdriver.Chrome(service=chrome_service)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Function to extract lexicon names from the current page
def extract_cat_names():
    cat_names = []
    lexique_categories = driver.find_elements(By.CLASS_NAME, 'CategoryTreeItem')
    for category in lexique_categories:
        name = category.text.strip().split('–')[0].strip()  # Split and take only the first part
        cat_names.append(name)
    return cat_names

# Function to extract lexicon terms from a category page
def extract_lexicon_terms():
    lexicon_terms = []
    terms = driver.find_elements(By.CSS_SELECTOR, '.CategoryTreeItem > a')
    for term in terms:
        lexicon_terms.append(term.text.strip())
    return lexicon_terms

# URL of the page
page_url = "https://fr.wiktionary.org/wiki/Catégorie:Lexiques_en_français"

# Navigate to the page
driver.get(page_url)

all_cat_names = []

# Extract lexicon names from all pages
while True:
    all_cat_names.extend(extract_cat_names())
    try:
        # Wait for the "Next page" link to be clickable
        next_page_link = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.LINK_TEXT, 'page suivante'))
        )
        next_page_link.click()  # Click on the "Next page" link
    except:
        break

lexicon_fr = []

# Function to extract lexicon terms from a category page
def extract_lexicon_terms():
    lexicon_terms = []
    terms = driver.find_elements(By.CSS_SELECTOR, '.mw-category-group > ul > li > a')
    for term in terms:
        lexicon_terms.append(term.text.strip())
    return lexicon_terms

# Navigate to the category's pages
for lexicon in all_cat_names:
    driver.get(f"https://fr.wiktionary.org/wiki/Cat%C3%A9gorie:{lexicon}")

# Extract lexicon terms and stock it the list
    lexicon_terms = extract_lexicon_terms()
    lexicon_fr.append({"catégorie": lexicon, "lexique" :lexicon_terms})
        
# Quit the driver
driver.quit()
```
